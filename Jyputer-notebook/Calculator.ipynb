# Python-Based calculator for basic AI math operation

```python
import numpy as np

# matirx input function

def input_matrix(name):
    print(f"\nEnter values for {name} (e.g. [1, 2, 3]): ")
    try:
        return np.array(eval(input()))
    except Exception as e:
        print("❌ Invaid input. Please enter in correct format.")
        return input_matrix(name)
    
# vector input function

def input_vector(name):
    print(f"\nEnter values for {name} (e.g. [1, 2, 3]): ")
    try:
        return np.array(eval(input()))
    except Exception as e:
        print("❌ Invaid input. Please enter in correct format.")
        return input_vector(name)
    
def matrix_multiply(A,B):
    try:
        return np.dot(A,B)
    except Exception as e:
        return f"❌ Error: {e}"
    
def vector_dot(v1,v2):
    try:
        return np.dot(v1,v2)
    except Exception as e:
        return f"❌ Error: {e}"

def sigmoid(x):
    return 1 / (1+np.exp(-x))


def relu(x):
    return np.maximum(0,x)

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()

def run_calculator():
    print("🧠 AI math calculator\n")

run_calculator()

while True:
    print("\nSelect an Operation:")
    print("\n1. Matrix Multiplication")
    print("\n2. Vector Dot Product")
    print("\n3. Sigmoid Function")
    print("\n4. Relu Function")
    print("\n5. Software Function")
    print("\n6. Exit")

    choice = input("Enter your choice (1-6)")
    if choice == "1":
        A = input_matrix("matrix A")
        B = input_matrix("matrix B")
        result = matrix_multiply(A,B)
        print("\n✅ Result:\n", result)

    elif choice == "2":
        v1 = input_vector("vector 1")
        v2 = input_vector("vector 2")
        result = vector_dot(v1,v2)
        print("\n✅Result:\n", result)

    elif choice == "3":
        v = int(input("input Vector"))
        result = sigmoid(v)
        print("\n✅ Sigmoid Output", result)

    elif choice == "4":
        v = int(input("Input Vector"))
        result = relu(v)
        print("\n✅ ReLU Output", result)

    elif choice == "5":
        v = input("Input vector")
        x = np.array(eval(v))
        result = softmax(x)
        print("\n✅ Softmax Output", result)

    elif choice == "6":
        print("\n👋 Exiting. Thank you!")
        break

    else:
        print("❌ Invaild choice. Please enter a number from 1 to 6.")


```
