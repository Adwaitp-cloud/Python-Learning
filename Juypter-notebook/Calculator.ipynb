# Python-Based calculator for basic AI math ops

```python
import numpy as np

# matirx input function

def input_matrix(name):
    print(f"\nEnter values for {name}(e.g. [[1,2],[3,4]])")
    try:
        return np.array(eval(input()))
    except Exception as e:
        print("❌ Lnvalid Lnput. Please enter the correct format")
        return input_matrix(name)
    
# Vector input function

def input_vector(name):
    print(f"\nEnter values for {name}(e.g. [[1,2],[3,4]])")
    try:
        return np.array(eval(input()))
    except Exception as e:
        print("❌ Lnvalid Lnput. Please enter the correct format")
        return input_vector(name)
    
```
